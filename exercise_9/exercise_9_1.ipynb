Write code to iterate over all feature classes in a geodatabase which are of the type ‘Point-Geometry’. For each feature class build a search cursor that iterates over the feature class. All rows that have the Value of ‘active’ in the field ‘status’ should be copied over into a new feature class named ‘active_assets’. 

```python
arcpy.env.workspace = r'C:\Users\Admin\Documents\Uni\PythonQGISArcGIS\Arcpy_Intro\exercise_arcpy_1.gdb'

# select all feature classes of the type point
fc_pt_list = arcpy.ListFeatureClasses(feature_type='Point')
print(fc_pt_list)

fields = ["SHAPE@", "status", "type"]

# iterate over the point feature classes
with arcpy.da.InsertCursor("active_assets", fields) as i_cursor:
    for fc in fc_pt_list:
        if fc == "active_assets": # don't copy rows from the target feature class
            continue
        print(fc)
        with arcpy.da.SearchCursor(fc, fields, where_clause="status = 'active'") as s_cursor:
            for row in s_cursor: # insert rows with status "active" into active_assets
                i_cursor.insertRow(row)
```
